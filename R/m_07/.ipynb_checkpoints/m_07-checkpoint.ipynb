```python
import os
import warnings
```

```python
dataset = "flights.csv"
# currnet file directory
wd = os.getcwd()
# check the dataset is in the directory
if not dataset in os.listdir(wd):
    warnings.warn("There is no input data", UserWarning)
```

```python
import pandas as pd # data processing, CSV file I/O
import numpy as np

# read the dataset
flights=pd.read_csv(os.path.join(wd, dataset))
```

```python
flights.head()
```

```python
flights = flights[flights["dest"] == "SFO"]
```

```python
flights.head()
```

```python
sample = flights.sample(n = 25)
```

```python
sample["arr_delay"].describe()
```

```python
sample["arr_delay"].quantile(0.98)
```

```python
sample.loc[sample["arr_delay"] <= 90, "proportion"] = "TRUE"
sample.loc[sample["arr_delay"] > 90, "proportion"] = "FALSE"
```

```python
sample["proportion"].value_counts(normalize=True)
```
