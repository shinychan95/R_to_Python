```python
import os
import warnings
```

```python
# currnet file directory
wd = os.getcwd()

# 데이터 셋의 이름이 워낙 길어서 책과 같은 방법으로 이름 저장
dataset = os.listdir(os.path.join(wd, "data", "fueleconomy"))[0]
```

```python
import pandas as pd # data processing, CSV file I/O
import numpy as np

# read the dataset
cars=pd.read_excel(os.path.join(wd, "data", "fueleconomy", dataset))
```

```python
cars.head()
```

### 데이터 정제 구간
cars에서 Mfr Name이 Toyota에 해당하는 것만 추려내고, Hierarchical clustering 에 필요한 데이터만을 따로 변수에 저장

```python
cars = cars[cars["Mfr Name"]=="Toyota"]
```

```python
labels = cars["Carline"]
```

```python
data = cars[["Eng Displ", "# Cyl", "# Gears", "City FE (Guide) - Conventional Fuel", "Hwy FE (Guide) - Conventional Fuel"]]
```

### distance matrix 만들기 위한 작업

```python
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist, jaccard
```

```python
matrix = pd.concat([data, labels], axis=1)
```

```python
matrix.head(5)
```

```python
matrix = matrix.groupby("Carline").mean()
```

```python
res = pdist(matrix, 'euclidean')
squareform(res)
distance = pd.DataFrame(squareform(res), index=matrix.index, columns= matrix.index)
```

```python
distance.head(5)
```

### scipy 를 이용한 계층 분석 모델 구현

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
# Calculate the linkage: mergings
mergings = linkage(data, method="complete")

# Plot the dendrogram, using varieties as labels
plt.figure(figsize=(16, 8))
dendrogram(mergings,
           labels = labels.as_matrix()
)
plt.show()
```

- 앞의 코드에서, linkage 함수에서 method 를 사용했다. 이에 대해서 알아보자.
- Hierachical clustering의 기본 원리는 두 클러스터 사이의 거리를 측정해서 거리가 가까운 클러스터끼리 묶는 방식이다.
- 그러면 두 클러스터의 거리를 측정할때 어디를 기준점으로 할것인가를 결정해야 하는데 다음 그림을 보자.
<img src="./image/cluster.PNG" width="500" height="500">
앞의 코드에서 사용한 complete linkage 방식은 두 클러스터상에서 가장 먼 거리를 이용해서 측정하는 방식이고 반대로  single linkage 방식은 두 클러스터에서 가장 가까운 거리를 사용하는 방식이다.

## 9.1.2 k-means

```python
dataset = "WorldCities.csv"

# check the dataset is in the directory
if not dataset in os.listdir(os.path.join(wd, "data")):
    warnings.warn("There is no input data", UserWarning)
```

```python
# read the dataset
WorldCities=pd.read_csv(os.path.join(wd, "data", dataset))
```

```python
WorldCities
```

```python
WorldCities = WorldCities.sort_values(by=["population"], ascending = False).iloc[:4000]
```

```python
WorldCities = WorldCities[["longitude", "latitude"]].reset_index(drop=True)
```

```python
WorldCities
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
# create model and prediction
model = KMeans(n_clusters=6, algorithm="auto")
model.fit(WorldCities)
```

```python
WorldCities["cluster"] = model.labels_
```

```python
plt.figure(figsize=(16, 8))
sns.scatterplot("longitude", "latitude", data=WorldCities, hue="cluster")
```

## 9.2 Dimension reduction
- 차원 축소를 다루는 것 같지만, 사실 SVD를 통해 어느 정도에서 의미있는 정보성을 찾는 것이다.

### SVD의 경우, 아래 링크를 살펴보면 정말 기하학적으로 가히 놀랄 정도로 이해할 수 있다.
- https://www.youtube.com/watch?v=cq5qlYtnLoY
    
**간략하게 SVD를 소개하면, 특이값 분해를 하는 것인데, 어떤 벡터를 정보량에 따라서 여러 Layer로 분해하는 것이다.**
- 인물 사진을 예를 들면,
    - 인물 사진의 경우 어마어마한 정보를 담고 있을텐데,
    - 사진을 흰 화면인 무에서 유로 형성해감에 있어, 정보량이 큰 Layer부터 작은 Layer의 역할을 살펴보면,
    - 정보량이 큰 Layer부터 차곡차곡 흰 화면에 겹쳐지면서, 
    - 초반에 정보량이 큰 Layer들을 더하면, 얼굴의 어느 정도 윤곽이 나타나고,
    - 작은 Layer들을 더하면서 사진이 조금씩 선명해지는 것으로 이해하면 된다.
    - 정보량이 큰 레이어들을 쌓아가면, 쌓기 전과 후과 명확히 달라지면서 확연히 얼굴로서 보이고,
    - 정보량이 작은 레이어들이 쌓일 때는, 쌓기 전과 후과 명확히 달라지지 않지만 선명해지는 것이다.
    
**자세하게 이해하기 위해서는 영상을 살펴보는 것이 아주 현명하다**

### 위 그래프를 보면 정말 아주 정보도 찾을 수 없음.
그래서 우선 일부에 혹시나 의미있는 정보값이 있지 않을까 찾아보기 (파이썬의 경우 position_jitter가 구현에 있어서 어려움이 큼)

<img src="./image/position_jitter.PNG" width="800" height="800">

```python
from sklearn.preprocessing import LabelEncoder

def dummyEncode(df):
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

```python
Votes = dummyEncode(Votes)
```

```python
Votes["bill"] = Votes["bill"] / Votes["bill"].max()
Votes["name"] = Votes["name"] / Votes["name"].max()
```

```python
Votes["bill"] = Votes["bill"].astype(np.float16)
Votes["name"] = Votes["name"].astype(np.float16)
```
