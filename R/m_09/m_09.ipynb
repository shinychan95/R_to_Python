cars에서 Mfr Name이 Toyota에 해당하는 것만 추려내고, Hierarchical clustering 에 필요한 데이터만을 따로 변수에 저장

### distance matrix 만들기 위한 작업

```python
from scipy.spatial.distance import squareform
from scipy.spatial.distance import pdist, jaccard
```

### scipy 를 이용한 계층 분석 모델 구현

```python
# Perform the necessary imports
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

```python
# Calculate the linkage: mergings
mergings = linkage(data, method="complete")

# Plot the dendrogram, using varieties as labels
plt.figure(figsize=(16, 8))
dendrogram(mergings,
           labels = labels.as_matrix()
)
plt.show()
```

- 앞의 코드에서, linkage 함수에서 method 를 사용했다. 이에 대해서 알아보자.
- Hierachical clustering의 기본 원리는 두 클러스터 사이의 거리를 측정해서 거리가 가까운 클러스터끼리 묶는 방식이다.
- 그러면 두 클러스터의 거리를 측정할때 어디를 기준점으로 할것인가를 결정해야 하는데 다음 그림을 보자.

### SVD의 경우, 아래 링크를 살펴보면 정말 기하학적으로 가히 놀랄 정도로 이해할 수 있다.

**간략하게 SVD를 소개하면, 특이값 분해를 하는 것인데, 어떤 벡터를 정보량에 따라서 여러 Layer로 분해하는 것이다.**
- 인물 사진을 예를 들면,
    - 인물 사진의 경우 어마어마한 정보를 담고 있을텐데,
    - 사진을 흰 화면인 무에서 유로 형성해감에 있어, 정보량이 큰 Layer부터 작은 Layer의 역할을 살펴보면,
    - 정보량이 큰 Layer부터 차곡차곡 흰 화면에 겹쳐지면서, 
    - 초반에 정보량이 큰 Layer들을 더하면, 얼굴의 어느 정도 윤곽이 나타나고,
    - 작은 Layer들을 더하면서 사진이 조금씩 선명해지는 것으로 이해하면 된다.
    - 정보량이 큰 레이어들을 쌓아가면, 쌓기 전과 후과 명확히 달라지면서 확연히 얼굴로서 보이고,
    - 정보량이 작은 레이어들이 쌓일 때는, 쌓기 전과 후과 명확히 달라지지 않지만 선명해지는 것이다.
    
**자세하게 이해하기 위해서는 영상을 살펴보는 것이 아주 현명하다**

**전체를 반반 나눠서 경향성을 파악하기 위함**

```python
from sklearn.preprocessing import LabelEncoder

def dummyEncode(df):
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

### 이제부터 진짜 SVD

```python
dataset = "Votes.csv"

# check the dataset is in the directory
if not dataset in os.listdir(os.path.join(wd, "data")):
    warnings.warn("There is no input data", UserWarning)
```

```python
# read the dataset
Votes=pd.read_csv(os.path.join(wd, "data", dataset))
```

```python
table = Votes.pivot(index="name", columns="bill", values="vote")
```

```python
table.shape
```

**scipy 라이브러리로 간단하게 특이값 분해를 한다**

```python
import scipy
```

```python
U, s, V = np.linalg.svd(table, full_matrices=True)
```

```python
U.shape
```

```python
s.shape
```

```python
V.shape
```

**특이값 분해 후 정보값의 크기에 있어서 상위 5개의 Sigma 값만을 이용해 기존 table의 근사치(A)를 구한다**

```python
S = np.zeros((table.shape[0], table.shape[1]))
S[:table.shape[0], :table.shape[0]] = np.diag(s)
n_component = 5
S = S[:, :n_component]
V = V[:n_component, :]
A = U.dot(S.dot(V))
print(A)
result = pd.DataFrame(A)
```

```python
result.head()
```

**하지만, 애초부터 hitmap의 경우 그림에서 특징조차 없었어서 사실 상위 시그마 값으로 근사 table을 구해도 hitmap을 통해서는 값을 찾을 수 없다**

```python
plt.figure(figsize=(16, 8))
sns.heatmap(result)
```

### U matrix에 대해서 클러스터링을 통한 정보성 분석

```python
U, s, V = np.linalg.svd(table, full_matrices=True)
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

M_U = pd.DataFrame(U[:, 0:5])

# create model and prediction
model = KMeans(n_clusters=6, algorithm="auto")
model.fit(M_U)

M_U["cluster"] = model.labels_

plt.figure(figsize=(16, 8))
sns.scatterplot(0, 1, data=M_U, hue="cluster")
```

### V matrix에 대해서 클러스터링을 통한 정보성 분석

```python
V.shape
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

M_V = pd.DataFrame(V.T[:, 0:5])

# create model and prediction
model = KMeans(n_clusters=16, algorithm="auto")
model.fit(M_V)

M_V["cluster"] = model.labels_

plt.figure(figsize=(16, 8))
sns.scatterplot(0, 1, data=M_V, hue="cluster")
```

### 책에서 말하길, 
**U에 대한 분석은 (0, 0)에 해당하는 average member로부터 세 가지 갈래로 나뉘는 것을 보인다고 말하며,**<br>
**V에 대한 분석은 (0, 0)에 해당하는 average member를 중심으로 대칭으로 진보와 보수로 나뉘는 것처럼 언급하고 있다.**
